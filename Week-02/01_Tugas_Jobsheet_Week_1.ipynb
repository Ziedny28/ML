## Nomor 2 - Lakukan proses encoding pada kolom "diagnosis".

- menampilkan 100 data teratas

```python
wbc.head(100)
```

- melakukan enkoding dengan menggunakan LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
le = LabelEncoder() #object dari LabelEncoder
wbc['diagnosis'] = le.fit_transform(wbc['diagnosis']) # proses encoding
wbc.head(100)
```

## Nomor 4 - Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

- melakuakan proses starttified split data untuk membuat data latih dan uji dengan rasio 80:20 menggunakan train_test_split

```python
# Split data
from sklearn.model_selection import train_test_split
x = wbc
y = wbc['diagnosis']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y)
```
